```python
#import librerie
import pandas as pd
import folium
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from folium.plugins import HeatMap
from folium.plugins import MarkerCluster 
```

Import dataset

```python
#import dataset from url
data=pd.read_excel("https://whc.unesco.org/en/list/xls/?2015")
data = pd.DataFrame(data)
data.head()
```

```python
#Q1
#Raggruppare per nazione e contare il numero di beni.
BenixNazioni = data.groupby("states_name_en")["name_en"].count().reset_index(name="Numero_beni")
#sort in ordine decrescente
BenixNazioni=BenixNazioni.sort_values(by="Numero_beni", ascending=False)
BenixNazioni
```

```python
# Q2
# Quale nazione ha più punti di interesse riconosciuti?
max_beni = BenixNazioni["Numero_beni"].idxmax()
max_state_beni= BenixNazioni.loc[max_beni, "states_name_en"]
print("Nazione con più punti di interesse riconosciuti:", max_state_beni,"con un numero di beni riconosciuti pari a", max(BenixNazioni.Numero_beni))
```

```python
#Q3
# Quale nazione ne ha meno punti di interesse riconosciuti?
indice_min = BenixNazioni.iloc[:, 1].idxmin()
riga_min = BenixNazioni.iloc[[indice_min]]
print("Di seguito le informazioni relative alla Nazione con meno punti di interesse riconosciuti:")
print(BenixNazioni[BenixNazioni["Numero_beni"]==1].nunique())
print("78 Nazioni hanno il minimo numero di beni che corriponde a", min(BenixNazioni.Numero_beni) )
```

```python
#Q6
#Rappresentare su mappa i siti
#contiamo i valori null e not null per latitudine e longitudine. 
#Creiamo un nuovo dataset copia che esclude i valori non nulli. 
print("Valori nulli latitudine:", data.latitude.isnull().sum())
print("Valori nulli longitudine:", data.longitude.isnull().sum())
print("Valori non nulli latitudine:", data['latitude'].notnull().sum())
print("Valori non nulli longitudine:", data['longitude'].notnull().sum())
```

```python
#Poichè soltanto un campo del dataframe ha valori di latitudine e longitudine nulli, li escludiamo.
data_mappe = data.loc[data.latitude.notnull() & data.longitude.notnull()].copy()
```

```python
#crezione mappa con la libreria folium per utilizzare e manipolare i dati Geopandas
def create_map( center = (0,0), zoom = 2 , width = 800, height=600 ) :
  mappa = folium.Map( location = center, zoom_start= zoom )
  figura = folium.Figure( width=width, height=height )
  return mappa.add_to(figura);
```

```python
#Si crea la mappa del mondo in cui sono indicati i siti dei beni riconosciuti in tutto il Mondo
mappa = create_map(center = (0,0),  zoom = 2, width="100%", height=400)
for index, row  in data_mappe.iterrows():
  mappa.add_child( folium.Marker(  location=(row.latitude, row.longitude), popup=row.name_en))

mappa
```

```python
#Per facilitare la visualizzazione si si utilizza la funzione MarkerCluster che permette di raggruppare insieme i marcatori che si sovrappongono in una mappa
#I popup sono raggruppati in cluster, e quando si fa clic su di essi, la mappa si ingrandisce in quel punto e mostra i marcatori singoli all'interno del cluster.
mc = MarkerCluster()
mappa= create_map(center=(30,15), zoom=2, width="100%", height=400)
for index, row in data_mappe.iterrows():
    mc.add_child(folium.Marker( location=(row.latitude, row.longitude), popup=row.name_en))

mappa.add_child(mc)
mappa
```

```python
#Q11
# Quante e quali sono le caterogie dei siti
Category=data[["category","states_name_en"]].loc[data.category.notnull() & data.states_name_en.notnull()].copy()
# NUMERO DI CATEGORIE
print(f"Le categorie sono", Category.category.value_counts().count(), "e sono:", Category.iloc[:, 0].value_counts().index.tolist())
```

```python
#Q12
# Qual è la categoria predominante di siti unesco in Italia, in Australia e in Vaticano?
# Funzione che definisce la categoria predominante dei siti Unesco in base alla Nazione scelta
def Siti(Nome_stato,dataset):
    Siti=dataset.loc[dataset.states_name_en == Nome_stato]
    Category_count=Siti.groupby(["category"]).count()
    idx = Category_count['states_name_en'].idxmax()
    print(f"La categoria predominante in {Nome_stato} è:",idx,"con un numero di siti=", Category_count["states_name_en"].max())
```

```python
# Categoria predominante dei siti Unesco in Italia
Siti("Italy",Category)
```

```python
# Categoria predominante dei siti Unesco in Australia
Siti("Australia",Category)
```

```python
# Categoria predominante dei siti Unesco nel Vaticano
Siti("Holy See,Italy",Category)
```
